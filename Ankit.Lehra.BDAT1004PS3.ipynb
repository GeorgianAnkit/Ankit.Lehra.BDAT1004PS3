```python
import pandas as pd

# Import the dataset
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')

# Mean age per occupation
print("\nMean Age Per Occupation:")
print(users.groupby('occupation')['age'].mean())

# Male ratio per occupation, sorted
male_ratio = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
print("\nMale Ratio Per Occupation:")
print(male_ratio.sort_values(ascending=False))

# Min and Max ages per occupation
print("\nMin and Max Ages Per Occupation:")
print(users.groupby('occupation')['age'].agg(['min', 'max']))

# Mean age per occupation and gender
print("\nMean Age Per Occupation and Gender:")
print(users.groupby(['occupation', 'gender'])['age'].mean())

# Gender percentage per occupation
gender_counts = users.groupby('occupation')['gender'].value_counts(normalize=True) * 100
print("\nGender Percentage Per Occupation:")
print(gender_counts)
```

```python
import pandas as pd
import datetime as dt

# Step 2
data = pd.read_csv('C:\\Users\\ankit\\Downloads\\Problem Set 3 Data Files\\wind.txt', delim_whitespace=True)

# Step 3
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=['Yr', 'Mo', 'Dy'])

# Step 4
def fix_year(date):
    if date.year > 2000:
        return date.replace(year=date.year-100)
    return date

data['Date'] = data['Date'].apply(fix_year)

# Step 5
data = data.set_index('Date')

# Step 6
missing_values = data.isnull().sum()
print(missing_values)

# Step 7
non_missing_values = data.count().sum()
print(non_missing_values)

# Step 8
mean_windspeed = data.mean().mean()
print(mean_windspeed)

# Step 9
loc_stats = data.describe().loc[['min', 'max', 'mean', 'std']].transpose()
print(loc_stats)

# Step 10
day_stats = data.agg(['min', 'max', 'mean', 'std'], axis=1)
print(day_stats)

# Step 11
january_windspeed = data[data.index.month == 1].mean()
print(january_windspeed)

# Step 12
yearly_data = data.resample('Y').mean()

# Step 13
monthly_data = data.resample('M').mean()

# Step 14
weekly_data = data.resample('W').mean()

# Step 15
weekly_stats_first_year = data.resample('W').agg(['min', 'max', 'mean', 'std']).iloc[1:53]
print(weekly_stats_first_year)
```

```python
# Step 1
import pandas as pd

# Step 2
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
# Step 3
chipo = pd.read_csv(url, sep="\t")

# Step 4
print(chipo.head(10))

# Step 5
print("Number of observations:", chipo.shape[0])

# Step 6
print("Number of columns:", chipo.shape[1])

# Step 7
print("Columns:", chipo.columns.tolist())

# Step 8
print("Index:", chipo.index)

# Step 9 & 10
most_ordered_item = chipo['item_name'].value_counts().idxmax()
number_of_most_ordered_item = chipo['item_name'].value_counts().max()
print(f"The most-ordered item: {most_ordered_item}, Number of times ordered: {number_of_most_ordered_item}")

# Step 11 & 12
most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()
number_of_most_ordered_choice_description = chipo['choice_description'].value_counts().max()
print(f"The most-ordered item in choice_description: {most_ordered_choice_description}, Number of times ordered: {number_of_most_ordered_choice_description}")

total_items_ordered = chipo['quantity'].sum()
print(f"Total items ordered: {total_items_ordered}")

# Step 13
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
print("Item price type after conversion:", chipo['item_price'].dtype)

# Step 14
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(f"Revenue: ${revenue}")

# Step 15
number_of_orders = chipo['order_id'].nunique()
print(f"Number of orders: {number_of_orders}")

# Step 16
average_revenue_per_order = revenue / number_of_orders
print(f"Average revenue per order: ${average_revenue_per_order}")

# Step 17
number_of_different_items_sold = chipo['item_name'].nunique()
print(f"Number of different items sold: {number_of_different_items_sold}")
```
